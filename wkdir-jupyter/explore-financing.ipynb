```python
out = []
for jsonfile in os.listdir("./"):
    if "financings-" in jsonfile and jsonfile.endswith(".json"):
        print(jsonfile)
        keyfile = jsonfile.replace(".json", "")
        ldata = json.loads(open(jsonfile).read())
        for k,v in ldata[0].items():
            if isinstance(v, list):
                for obj in v:
                    for k2 in obj:
                        toinsert = f"\"{keyfile}\" -> \"{k}.{k2}\""
                        if not toinsert in out:
                            out.append(toinsert)
            else:
                toinsert = f"\"{keyfile}\" -> \"{k}\""
                if not toinsert in out:
                    out.append(toinsert)
```

```python
print("\n".join(out))
```

```python
db = {}
for o in out:
    o = o.replace('"', '')
    k, v = o.split(' -> ')
    if not v in db:
        db[v] = {}
    db[v][k] = 1
```

```python
ff = [
    'customerId',
    'contractNumber',
    'contractId',
    'organizationId',
    'organizationName',
    'productName',
    'productType',
    'warrantyType',
    'warrantySubType',
    'warrantyAmount',
    'brandName',
    'contractAmount',
    'totalNumberOfInstalments',
    'paidInstalments',
    'contractOutstandingBalance',
]
for k in ff:
    print(k)
    for k2 in db[k]:
        print("\t", k2)
```
