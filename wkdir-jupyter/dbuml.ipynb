```python
colmap = {}
for k,v in elems.items():
    k = k.replace("-", "_")
    print(f"Table {k} {{")
    for k2 in v[0]:
        print("    ", k2, "varchar")
        if not k2 in colmap:
            colmap[k2] = {}
        colmap[k2][k] = 1
    print("}")
```

```python
ref = []
for k in colmap:
    arr = list(colmap[k].keys())
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            print(i,j)
            r = f"Ref: {arr[i]}.{k} > {arr[j]}.{k}"
            ref.append(r)
```

```python
print("\n".join(ref))
```
